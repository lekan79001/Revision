```python
def social_media_prices(quantity):
    
    
    if 0 <= quantity <= 5:
        price = 30000
    elif 5> quantity <= 10:
        price = 45000
    elif 10> quantity <=20:
        price = 60000
    else:
        return "OUT OF RANGE"
    return price
```

```python
social_media_prices(15)
```
